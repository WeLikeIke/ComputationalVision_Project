```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

%matplotlib notebook
```

```python
#Loading the data
F3D = np.load('PCA.npy')
f3d = np.load('TSNE.npy')
labels = np.load('labels.npy')

print(np.shape(F3D),np.shape(f3d),np.shape(labels))
```

```python
#Setting up a color for each class
color_list = ['red', 'blue', 'green', 'yellow', 'black', 'brown', 'cyan', 'pink']
color = []
for i in labels:
    color.append(color_list[i])
```

```python
#Interactive 3D plot
fig = plt.figure(figsize=plt.figaspect(0.5))

ax = fig.add_subplot(1,2,1, projection='3d')
for i,f in enumerate(F3D):
    ax.scatter3D(f[0],f[1],f[2], c=color[i])
    
axt = fig.add_subplot(1,2,2, projection='3d')
for i,f in enumerate(f3d):
    axt.scatter3D(f[0],f[1],f[2], c=color[i])
```
